### 2.3 Встроенные типы последовательностей: `str`, `list`, `tuple`

#### 1) Изменение строки `S = 'spam'` на `'slam'` только с помощью нарезания и конкатенации

Согласно лекции, строка — это **неизменяемая** последовательность, поэтому она не меняется «на месте».  
Нужная строка создаётся как новый объект.


```python

S = 'spam'
S_new = S[:1] + 'l' + S[2:]
S, S_new
```


Индексацию и конкатенацию тоже можно использовать, если собрать **новую** строку из частей.  
А вот присваивание по индексу для строки невозможно, потому что строки неизменяемы.


```python

S = 'spam'
S_new_index = S[0] + 'l' + S[2] + S[3]
S_new_index
```

```python

# Присваивание по индексу для строки вызывает ошибку:
try:
    S = 'spam'
    S[1] = 'l'
except TypeError as e:
    print(type(e).__name__ + ':', e)
```


Для списка похожая операция возможна и через создание нового списка, и через изменение по индексу, потому что список изменяем.  
Для кортежа — только через создание нового кортежа, потому что кортеж неизменяем.


```python

# Список
L = list('spam')
L_new = L[:1] + ['l'] + L[2:]
L_index = list('spam')
L_index[1] = 'l'

# Кортеж
T = tuple('spam')
T_new = T[:1] + ('l',) + T[2:]

print('L_new  =', L_new)
print('L_index=', L_index)
print('T_new  =', T_new)
```

#### 5) Три различных списковых метода для добавления элемента


```python

L1 = [1, 2, 3]
L1.append(4)

L2 = [1, 2, 3]
L2.extend([4])

L3 = [1, 2, 3]
L3.insert(len(L3), 4)

print('append ->', L1)
print('extend ->', L2)
print('insert ->', L3)
```

#### 6) Перестановка элементов списка в обратном порядке с помощью операции и метода

Для списка можно использовать:
- операцию расширенного среза `[::-1]`;
- метод `reverse()`.

Для строки и кортежа можно получить обратный порядок срезом `[::-1]`,  
но метода `reverse()` у них нет.


```python

L = [1, 2, 3, 4]
L_slice = L[::-1]

L_method = [1, 2, 3, 4]
L_method.reverse()

S = 'spam'
T = (1, 2, 3, 4)

print('Список, срез :', L_slice)
print('Список, метод:', L_method)
print('Строка       :', S[::-1])
print('Кортеж       :', T[::-1])

for obj in ('spam', (1, 2, 3)):
    try:
        obj.reverse()
    except AttributeError as e:
        print(type(obj).__name__, '->', type(e).__name__ + ':', e)
```


#### 10) Четыре способа копирования списка. Отличие `copy()` и `deepcopy()`


```python

import copy

original = [[1, 2], 3, 4]

copy_1 = original[:]              # срез
copy_2 = list(original)           # конструктор
copy_3 = original.copy()          # метод copy
copy_4 = copy.deepcopy(original)  # глубокая копия

copy_1[1] = 30
copy_2[2] = 40
copy_3[0][0] = 100
copy_4[0][1] = 200

print('original =', original)
print('copy_1   =', copy_1)
print('copy_2   =', copy_2)
print('copy_3   =', copy_3)
print('copy_4   =', copy_4)
```


**Комментарий.**  
Первые три способа создают **поверхностную копию**: внешний список новый, но вложенные объекты остаются общими.  
Поэтому изменение `copy_3[0][0]` повлияло и на `original`.  

`deepcopy()` создаёт копии и вложенных объектов, поэтому изменение `copy_4` не меняет исходный список.


### 2.12 Файловый объект

#### 16) Пример использования `open`, `close`, `read`, `readline`, `readlines`, `write`, `writelines`


```python

file_name = 'lab2_sample.txt'

# запись в файл
f = open(file_name, 'w', encoding='utf-8')
f.write('Первая строка\n')
f.writelines(['Вторая строка\n', 'Третья строка\n'])
f.close()

# чтение всего файла
f = open(file_name, 'r', encoding='utf-8')
all_text = f.read()
f.close()

# чтение одной строки
f = open(file_name, 'r', encoding='utf-8')
first_line = f.readline()
f.close()

# чтение всех строк в список
f = open(file_name, 'r', encoding='utf-8')
all_lines = f.readlines()
f.close()

print('read() ->')
print(all_text)
print('readline() ->', first_line)
print('readlines() ->', all_lines)
```

## Задание 2.3. Фрагменты кода с ошибками: причина и исправление

### Ошибка 1
**Исходный фрагмент:** ``str1[10]``

**Причина:** Ошибка `IndexError`: обращение к индексу за пределами строки.

**Исправленный пример:**

```python
str1 = 'String'
print(str1[-1])
```

### Ошибка 2
**Исходный фрагмент:** ``[1] + '1'``

**Причина:** Ошибка `TypeError`: операция `+` для последовательностей требует объекты одного типа.

**Исправленный пример:**

```python
print([1] + ['1'])
```

### Ошибка 3
**Исходный фрагмент:** ``dict1['two']``

**Причина:** Ошибка `KeyError`: в словаре нет ключа `'two'`.

**Исправленный пример:**

```python
dict1 = {'one': 1}
print(dict1.get('two', 'ключ отсутствует'))
```

### Ошибка 4
**Исходный фрагмент:** ``t = [1,2], {1:2}, {1,2}; t[0] = 0``

**Причина:** Ошибка `TypeError`: `t` — кортеж, а кортеж неизменяем.

**Исправленный пример:**

```python
t = ([1, 2], {1: 2}, {1, 2})
print((0,) + t[1:])
```

### Ошибка 5
**Исходный фрагмент:** ``{1, 'str', (1,)}.add([1])``

**Причина:** Ошибка `TypeError`: список не является хешируемым объектом, поэтому его нельзя добавить в множество.

**Исправленный пример:**

```python
s = {1, 'str', (1,)}
s.add((1,))
print(s)
```

### Ошибка 6
**Исходный фрагмент:** ``dict1['three']` после `dict1.get('three')``

**Причина:** `get()` не вызывает ошибку и возвращает `None`, но обращение `dict1['three']` приводит к `KeyError`, если ключа нет.

**Исправленный пример:**

```python
dict1 = {'one': 1}
print(dict1.get('three', 'ключ отсутствует'))
```

### Ошибка 7
**Исходный фрагмент:** ``s.index('-')``

**Причина:** Ошибка `ValueError`: подстрока `'-'` не найдена.

**Исправленный пример:**

```python
s = 'string'
print(s.find('-'))
```

### Ошибка 8
**Исходный фрагмент:** ``a, d` после создания только `a``

**Причина:** Ошибка `NameError`: переменная `d` не была определена.

**Исправленный пример:**

```python
a = 1
d = 2
print(a, d)
```

### Ошибка 9
**Исходный фрагмент:** ``abs = 5; abs(-1)``

**Причина:** Ошибка `TypeError`: имя `abs` переопределено числом, поэтому встроенная функция больше недоступна под этим именем.

**Исправленный пример:**

```python
abs_value = 5
print(abs(-1))
print(abs_value)
```

### Глава 4 — вопрос 3. Что означает неизменяемость, и какие типы считаются неизменяемыми?


Неизменяемый объект нельзя изменить после создания. В материалах по главе 4 к таким типам отнесены **числа, строки и кортежи**.  
Это означает, что операция изменения создаёт новый объект, а не меняет существующий «на месте».


```python

s = 'spam'
new_s = 'l' + s[1:]
print('Исходная строка:', s)
print('Новая строка   :', new_s)

t = (1, 2, 3)
new_t = (0,) + t[1:]
print('Исходный кортеж:', t)
print('Новый кортеж   :', new_t)
```

### Глава 7 — вопрос 4. Как можно изменить строку в Python?


Строку нельзя изменить по индексу, потому что тип `str` неизменяем.  
Нужно создать новую строку — например, с помощью срезов, конкатенации или метода `replace`.


```python

s = 'spam'
s1 = s[:1] + 'l' + s[2:]
s2 = s.replace('p', 'l')

print('Через срезы   :', s1)
print('Через replace :', s2)
```

### Глава 8 — вопрос 4. Назовите операции, изменяющие словарь


Словарь изменяется в основном через:
- присваивание по ключу;
- удаление `del`;
- метод `update`;
- метод `pop`.


```python

d = {'a': 1}
d['b'] = 2           # присваивание
d.update({'c': 3})   # update
removed = d.pop('a') # pop
del d['b']           # del

print('Словарь после изменений:', d)
print('Удалённое значение      :', removed)
```

### Глава 9 — вопрос 4. Как сохранить объекты Python в файле без перевода в строки?


В ответах к главе 9 для этого рекомендуется модуль `pickle`.  
Он позволяет сохранять объекты Python в бинарном виде.


```python

import pickle

data = {'name': 'Sergey', 'scores': [9, 10, 10]}
pickle_file = 'lab2_pickle.bin'

with open(pickle_file, 'wb') as f:
    pickle.dump(data, f)

with open(pickle_file, 'rb') as f:
    loaded = pickle.load(f)

print('Исходный объект :', data)
print('Загруженный объект:', loaded)
```
