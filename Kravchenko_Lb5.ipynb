# Основные определения

Треугольник Рело — область пересечения трех кругов радиуса \(r\) с центрами в вершинах равностороннего треугольника со стороной \(r\).

Так как треугольник Рело является фигурой постоянной ширины, его можно вписать в квадрат со стороной, равной ширине треугольника Рело.

Поступательное движение фигуры — движение, при котором траектории движения всех точек фигуры одинаковы.

Вращательное движение фигуры — движение, при котором траектории точек фигуры представляют собой окружности или дуги окружностей с центрами, лежащими в одной точке.

Качение треугольника Рело по квадрату происходит за счет одновременного поступательного и вращательного движения.

# Задание 5.1. Начальное положение треугольника Рело и квадрата



````python
n = 3                         # количество вершин треугольника Рело
center = np.array([0., 0.])   # центр треугольника Рело
r = 10                        # ширина треугольника Рело
N = 100                       # количество точек для описания стороны треугольника Рело
````

````python
relo_matrix = relo.regular_polygon_Relo(r=r, center=center, N=N)
````

````python
square = np.array([
    [-r / 2, -r / 2],
    [ r / 2, -r / 2],
    [ r / 2,  r / 2],
    [-r / 2,  r / 2],
    [-r / 2, -r / 2]
])
````

````python
plt.figure()
plt.plot(square[:, 0], square[:, 1], 'b-')
plt.plot(relo_matrix[:, 0], relo_matrix[:, 1], 'r-')
plt.plot(center[0], center[1], 'k.', markersize=5)
plt.axis('equal');
````

Треугольник Рело нужно сдвинуть влево так, чтобы он оказался вписанным в квадрат.

В исходном положении правая вершина треугольника находится дальше правой стороны квадрата.  
Правая сторона квадрата имеет координату:

\[
x=\frac r2.
\]

Максимальная \(x\)-координата треугольника Рело равна:

\[
\max(x_{\text{Relo}}).
\]

Поэтому величина горизонтального сдвига:

\[
shift_x=\max(x_{\text{Relo}})-\frac r2.
\]

Вектор сдвига:

\[
shift=(shift_x,0).
\]

Чтобы выполнить сдвиг влево, из координат треугольника вычитается `shift`.

````python
shift = np.array([np.max(relo_matrix[:, 0]) - r / 2, 0.])
shift
````

````python
relo_matrix_shifted = relo_matrix - shift
center = center - shift
````

````python
plt.figure()
plt.plot(square[:, 0], square[:, 1], 'b-')
plt.plot(relo_matrix_shifted[:, 0], relo_matrix_shifted[:, 1], 'r-')
plt.plot(center[0], center[1], 'k.', markersize=5)
plt.axis('equal');
````

# Задание 5.2. Анимация поступательного движения треугольника Рело по окружности

.

````python
N_frames = 100
````

Создадим матрицу `center_frame` из `N_frames` строк и двух столбцов.  
Она содержит координаты точек окружности радиуса `shift[0]` с центром в начале координат.

Обход выполняется от \(-\pi\) до \(\pi\), потому что в первом кадре центр должен находиться слева от начала координат:

\[
(\,shift[0]\cos(-\pi),\ shift[0]\sin(-\pi)\,) = (-shift[0],0).
\]

Именно это положение соответствует сдвигу треугольника влево относительно исходной позиции.

````python
t_center = np.linspace(-np.pi, np.pi, N_frames)
center_frame = shift[0] * np.transpose([np.cos(t_center), np.sin(t_center)])
````

Изобразим первый кадр анимации поступательного движения.

````python
frame = 0
relo_matrix_shifted = relo_matrix + center_frame[frame]

plt.figure()
plt.plot(square[:, 0], square[:, 1], 'b-')
plt.plot(relo_matrix_shifted[:, 0], relo_matrix_shifted[:, 1], 'r-')
plt.plot(center_frame[frame, 0], center_frame[frame, 1], 'r.', markersize=10)
plt.plot(center_frame[:, 0], center_frame[:, 1], 'k--')
plt.axis('equal');
````

## Пользовательская функция `draw_frame(r, N_frames, frame)`

Нужно написать функцию трех аргументов, которая изображает кадр с номером `frame` при поступательном движении центра треугольника Рело по окружности против часовой стрелки.

Требования:

- `r` задает ширину треугольника Рело;
- `N_frames` задает количество кадров;
- `frame` задает номер кадра, по умолчанию `frame=0`;
- функция выполняет визуализацию и не возвращает объекты;
- нужно указать аннотации типов;
- нужно оформить строки документации;
- переменные из глобальной области видимости в теле функции использовать нельзя.

````python
def draw_frame(r: float, N_frames: int, frame: int = 0) -> None:

    assert r > 0, 'r should be positive'
    assert isinstance(N_frames, int), 'N_frames should be integer'
    assert N_frames > 0, 'N_frames should be positive'
    assert isinstance(frame, int), 'frame should be integer'
    assert 0 <= frame < N_frames, 'frame should be in range from 0 to N_frames-1'

    n_local = 3
    center_local = np.array([0., 0.])
    N_local = 100

    relo_matrix_local = relo.regular_polygon_Relo(
        n=n_local,
        r=r,
        center=center_local,
        N=N_local
    )

    square_local = np.array([
        [-r / 2, -r / 2],
        [ r / 2, -r / 2],
        [ r / 2,  r / 2],
        [-r / 2,  r / 2],
        [-r / 2, -r / 2]
    ])

    shift_local = np.array([np.max(relo_matrix_local[:, 0]) - r / 2, 0.])

    t_center_local = np.linspace(-np.pi, np.pi, N_frames)
    center_frame_local = shift_local[0] * np.transpose([
        np.cos(t_center_local),
        np.sin(t_center_local)
    ])

    relo_matrix_changed = relo_matrix_local + center_frame_local[frame]

    plt.plot(square_local[:, 0], square_local[:, 1], 'b-')
    plt.plot(relo_matrix_changed[:, 0], relo_matrix_changed[:, 1], 'r-')
    plt.plot(center_frame_local[frame, 0], center_frame_local[frame, 1], 'r.', markersize=10)
    plt.plot(center_frame_local[:, 0], center_frame_local[:, 1], 'k--')
    plt.axis('equal')
````

````python
plt.figure()

N_frames = 100

plt.subplot(1, 3, 1)
draw_frame(r=r, N_frames=N_frames)

plt.subplot(1, 3, 2)
draw_frame(r=r, N_frames=N_frames, frame=20)

plt.subplot(1, 3, 3)
draw_frame(r=r, N_frames=N_frames, frame=40)
````

# Задание 5.3. Анимация качения треугольника Рело



Определим функцию `rotate_matrix(theta)` для создания матрицы поворота по часовой стрелке на угол `theta`.

Матрица поворота по часовой стрелке:

\[
\begin{pmatrix}
\cos\theta & \sin\theta\\
-\sin\theta & \cos\theta
\end{pmatrix}.
\]

````python
rotate_matrix = lambda theta: np.array([
    [np.cos(theta), np.sin(theta)],
    [-np.sin(theta), np.cos(theta)]
])

rotate_matrix(np.pi)
````

Повернем треугольник Рело с центром в начале координат на угол \(\pi/3\) по часовой стрелке.

````python
theta = np.pi / 3
matrix = rotate_matrix(theta)
relo_matrix_rotated = np.array([np.dot(matrix, row) for row in relo_matrix])
````


 альтернативная реализация кода:

```python
np.array([np.dot(matrix, row) for row in relo_matrix])
```

с использованием матричного умножения `matrix1 @ matrix2`.

Так как точки треугольника хранятся строками в матрице `relo_matrix`, можно использовать транспонирование:

\[
(matrix @ relo\_matrix^T)^T.
\]

Код:

```python
relo_matrix_rotated_alt = (matrix @ relo_matrix.T).T
```

````python
relo_matrix_rotated_alt = (matrix @ relo_matrix.T).T
np.allclose(relo_matrix_rotated, relo_matrix_rotated_alt)
````

````python
plt.figure()
plt.plot(relo_matrix[:, 0], relo_matrix[:, 1], 'r-')
plt.plot(relo_matrix[0, 0], relo_matrix[0, 1], 'r.', markersize=20)
plt.plot(relo_matrix_rotated[:, 0], relo_matrix_rotated[:, 1], 'g--')
plt.plot(relo_matrix_rotated[0, 0], relo_matrix_rotated[0, 1], 'g.', markersize=20)
plt.axis('equal');
````

Определим массив `theta` со значениями угла поворота при вращательном движении, согласованный с полным оборотом центра за `N_frames` кадров.

\[
2\pi \quad \longleftrightarrow \quad \frac{2\pi}{3}.
\]

````python
theta = np.linspace(0, 2 * np.pi / 3, N_frames)
````

Для совмещения вращательного и поступательного движения в кадре `frame`:

1. сначала поворачиваем треугольник Рело с центром в начале координат на угол `theta[frame]`;
2. затем переносим треугольник Рело на вектор `center_frame[frame]`.

Сначала нужно делать поворот, а потом перенос, потому что вращение должно происходить относительно собственного центра треугольника.  
Если сначала перенести фигуру, то поворот будет выполнен относительно начала координат, а не относительно центра треугольника, и движение станет неправильным.

````python
frame = 20

matrix = rotate_matrix(theta[frame])

relo_matrix_changed = (
    np.array([np.dot(matrix, row) for row in relo_matrix])
    + center_frame[frame]
)

plt.figure()
plt.plot(square[:, 0], square[:, 1], 'b-')
plt.plot(relo_matrix_changed[:, 0], relo_matrix_changed[:, 1], 'r-')
plt.plot(center_frame[frame, 0], center_frame[frame, 1], 'r.', markersize=10)
plt.plot(center_frame[:, 0], center_frame[:, 1], 'k--')
plt.axis('equal');
````

## Новая версия функции `draw_frame(r, N_frames, frame)`

Теперь функция изображает кадр качения треугольника Рело по квадрату с учетом:

1. поступательного движения центра по окружности против часовой стрелки;
2. вращательного движения треугольника по часовой стрелке.

Переменные из глобальной области видимости внутри функции использовать нельзя.

````python
def draw_frame(r: float, N_frames: int, frame: int = 0) -> None:

    assert r > 0, 'r should be positive'
    assert isinstance(N_frames, int), 'N_frames should be integer'
    assert N_frames > 0, 'N_frames should be positive'
    assert isinstance(frame, int), 'frame should be integer'
    assert 0 <= frame < N_frames, 'frame should be in range from 0 to N_frames-1'

    n_local = 3
    center_local = np.array([0., 0.])
    N_local = 100

    relo_matrix_local = relo.regular_polygon_Relo(
        n=n_local,
        center=center_local,
        r=r,
        N=N_local
    )

    square_local = np.array([
        [-r / 2, -r / 2],
        [ r / 2, -r / 2],
        [ r / 2,  r / 2],
        [-r / 2,  r / 2],
        [-r / 2, -r / 2]
    ])

    shift_local = np.array([np.max(relo_matrix_local[:, 0]) - r / 2, 0.])

    t_center_local = np.linspace(-np.pi, np.pi, N_frames)
    center_frame_local = shift_local[0] * np.transpose([
        np.cos(t_center_local),
        np.sin(t_center_local)
    ])

    theta_local = np.linspace(0, 2 * np.pi / 3, N_frames)
    rotate_matrix_local = lambda theta: np.array([
        [np.cos(theta), np.sin(theta)],
        [-np.sin(theta), np.cos(theta)]
    ])

    matrix_local = rotate_matrix_local(theta_local[frame])

    relo_matrix_changed = (
        np.array([np.dot(matrix_local, row) for row in relo_matrix_local])
        + center_frame_local[frame]
    )

    plt.plot(square_local[:, 0], square_local[:, 1], 'b-')
    plt.plot(relo_matrix_changed[:, 0], relo_matrix_changed[:, 1], 'r-')
    plt.plot(center_frame_local[frame, 0], center_frame_local[frame, 1], 'r.', markersize=10)
    plt.plot(center_frame_local[:, 0], center_frame_local[:, 1], 'k--')
    plt.axis('equal')
````
