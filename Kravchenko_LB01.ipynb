```python
import math
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as const
import sympy
```

```python
h_start = 1.0
h_end = 3.0E+0
T = 3.0
s_end = 5.0e0
```

```python
g = const.g
g, const.unit('standard acceleration of gravity')
```

```python
step = 0.01
t_list = [0 + i*step for i in range(int((T)/step))]
t_array = np.arange(0,T,step)
```

```python
v0_x = s_end/T
v0_y = ((h_end - h_start)+g/2*T**2)/T
apl = math.atan(v0_y/v0_x)
math.degrees(apl)
```

f строки ( просто пишешь текст, но можно вставлять переменные за счет фигурных скобок)

```python
x_list = [v0_x*t for t in t_list]
y_list = [h_start+v0_y*t-g/2*t**2  for t in t_list]
```

```python
x_array = v0_x*t_array
y_array = h_start + v0_y*t_array - g/2*t_array**2
```

```python
plt.figure()
plt.plot(x_list,y_list,'b-',lw=5)
plt.title('Вертикальное перемещение тела')
plt.xlabel('x (m)'); plt.ylabel('y (m)') # символ ; позволяет записывать несколь
plt.axis([0, s_end, 0, h_end+11])
```

```python
import cmath
def solver(ac, b, c):
    if ac == 0:
        if b == 0:
            return 'your equation kinda doesnt exist)', None
        else:
            return 'its a linear equation and its only root is ', -c/b, None
    D = (b**2) - (4*ac*c)
    sl1 = (-b + cmath.sqrt(D)) / (2*ac)
    sl2 = (-b - cmath.sqrt(D)) / (2*ac)
    return sl1, sl2
```

```python
print(solver(0,-6, 5))
```
